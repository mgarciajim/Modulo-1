### ¡BIENVENIDX AL JUEGO CAP-IT! ###

- Cada jugador cuando empieza el juego tiene 3 vidas.
- Hay 3 niveles: FÁCIL, MEDIO Y DIFÍCIL.
- Cada tres preguntas sube la dificultad.
- Si falla una respuesta, pasas a la pregunta siguiente.
- Son 9 preguntas en total.
- Cada fallo resta una vida, al terminar tus vidas, sales del juego.
- A cada respuesta el programa te devuelve "Correcto" o "Incorrecto, la capital correcta es XXX".
- Si pierdes todas tus vidas, salta un mensaje que dice "Has perdido todas tus vidas, (Nombre jugador)".


```python
# Juego preguntas de geografía:
# El usuario pone su nombre
# Crear diccionario con los paises europeos como claves y las capitales como valor. El diccionario está agrupado por tres distintos niveles.

capitales_con_paises = {
    
"facil" : {"Alemania": {"pais": "Alemania", "capital": "Berlin"},
"Francia": {"pais": "Francia", "capital": "Paris"},
"Italia": {"pais": "Italia", "capital": "Roma"},
"España": {"pais": "España", "capital": "Madrid"},
"Portugal": {"pais": "Portugal", "capital": "Lisboa"},
"Reino Unido": {"pais": "Reino Unido", "capital": "Londres"},
"Grecia": {"pais": "Grecia", "capital": "Atenas"},
"Países Bajos": {"pais": "Países Bajos", "capital": "Amsterdam"},
"Bélgica": {"pais": "Bélgica", "capital": "Bruselas"},
"Suiza": {"pais": "Suiza", "capital": "Berna"},
"Austria": {"pais": "Austria", "capital": "Viena"},
"Suecia": {"pais": "Suecia", "capital": "Estocolmo"}},
    
"medio" : {"Noruega": {"pais": "Noruega", "capital": "Oslo"},
"Dinamarca": {"pais": "Dinamarca", "capital": "Copenhague"},
"Finlandia": {"pais": "Finlandia", "capital": "Helsinki"},
"Polonia": {"pais": "Polonia", "capital": "Varsovia"},
"República Checa": {"pais": "República Checa", "capital": "Praga"},
"Irlanda": {"pais": "Irlanda", "capital": "Dublin"},
"Hungría": {"pais": "Hungría", "capital": "Budapest"},
"Rumanía": {"pais": "Rumanía", "capital": "Bucarest"},
"Bulgaria": {"pais": "Bulgaria", "capital": "Sofia"},
"Eslovaquia": {"pais": "Eslovaquia", "capital": "Bratislava"},
"Eslovenia": {"pais": "Eslovenia", "capital": "Liubliana"},
"Croacia": {"pais": "Croacia", "capital": "Zagreb"},
"Estonia": {"pais": "Estonia", "capital": "Tallin"}},
               
"dificil" : {"Letonia": {"pais": "Letonia", "capital": "Riga"},
"Lituania": {"pais": "Lituania", "capital": "Vilna"},
"Islandia": {"pais": "Islandia", "capital": "Reikiavik"},
"Malta": {"pais": "Malta", "capital": "La Valeta"},
"Luxemburgo": {"pais": "Luxemburgo", "capital": "Luxemburgo"},
"Ucrania": {"pais": "Ucrania", "capital": "Kiev"},
"Serbia": {"pais": "Serbia", "capital": "Belgrado"},
"Bosnia y Herzegovina": {"pais": "Bosnia y Herzegovina", "capital": "Sarajevo"},
"Moldavia": {"pais": "Moldavia", "capital": "Chisinau"},
"Liechtenstein": {"pais": "Liechtenstein", "capital": "Vaduz"},
"San Marino": {"pais": "San Marino", "capital": "San Marino"},
"Montenegro": {"pais": "Montenegro", "capital": "Podgorica"},
"Macedonia del Norte": {"pais": "Macedonia del Norte", "capital": "Skopie"}}
}
```

```python
# Elegimos valores aleatorios del diccionario:
import random


# Definimos función del juego:   
def juego():     
 
 # Creamos unas variables:
    vidas = 3
    preguntas_totales = 9
    level = 1
    preguntas_hechas = 0
    puntuacion = 0


# Generamos este input para que el usuario ingrese sus datos y tenga las instrucciones:

    nombre_jugador = input(f"¿Cual es tu nombre?")
    print(f"¡Bienvenido/a, {nombre_jugador}, al emocionante juego de las capitales!.\nTienes 3 vidas para responder correctamente a 9 preguntas.\nCada tres respuestas correctas, la dificultad aumentará.\n¡Buena suerte, que comience el juego!")
   
        
    niveles = ["facil", "medio", "dificil"]

# Creamos un bucle para que se ejecute indefinidamente mientras haya vidas:

    while preguntas_hechas < preguntas_totales and vidas > 0:

     # Como la lista empieza en 0 creamos esta variable para que el juego arranque en nivel fácil:   
        nivel_actual = niveles[level - 1]
            
     # Seleccionamos claves al azar del diccionario:
        pais = random.choice(list(capitales_con_paises[nivel_actual].keys()))
        capital_correcta = capitales_con_paises[nivel_actual][pais]['capital']

      # Creamos variable para guardar la respuesta del usuario:
        respuesta = input(f"¿Cual es la capital de {pais}?").strip()
                
      # Condición minúscula:            
        if respuesta.lower() == capital_correcta.lower():
            puntuacion += 1
            print(f"¡Correcto!")
            
    # Condición si hay fallo:
        else:
            vidas -= 1
            print(f"Incorrecto, la capital de {pais} es {capital_correcta}.")
            
        
        preguntas_hechas += 1

        # Cambio de nivel:
            
        if puntuacion % 3 == 0 and level < len(niveles):
            level += 1

        # Mensajes según puntuación:

        if puntuacion == 2:
            print(f'Eres como Don Quijote, {nombre_jugador}.')
        
        elif puntuacion == 4:
            print('¡Estás llegando a la frontera!')

        elif puntuacion == 6:
            print(f'¡Conquistando como Napoleón, {nombre_jugador}!')

        elif puntuacion == 8:
            print(f'Enhorabuena {nombre_jugador}, o mejor dicho, ¡Willy Fog!')

        elif puntuacion == 9:
            print('¡Bravo!¡Hasta el infinito y más allá!')    

    # Condición de pérdida de todas las vidas:

    if vidas == 0:

        print(f"\n¡Has perdido todas tus vidas, {nombre_jugador}!")
    
            
      # Mensaje final:

    else:

        print(f'¡Felicidades! Has terminado el juego con {vidas} vidas y {puntuacion} preguntas acertadas.')


   
```

```python
juego()
```
